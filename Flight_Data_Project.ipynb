```
import pandas as pd
import io
from google.colab import files
uploaded = files.upload()


```

Sort value in descending order

```
flightdata = pd.read_csv(io.BytesIO(uploaded['flight_details.csv']),  parse_dates=['flight_date'])
flightdata.sort_values(by='arr_delay', ascending=False)

```

**Questions and Solutions**

A. How often did delays occur from January 1st-15th?

```
"""Using lambdas function to find how often delays occur from January 1st-15th"""
flightdata['delayed'] = flightdata['arr_delay'].apply(lambda x: x > 0)
delays = flightdata['delayed'].value_counts()
print (delays)
print('Delays occured', delays[1], 'times between January 1st and 15th, 2015.')
```

B. Find aggregate trends and the percentage of flight delays

	i.)	Number of flights each carrier flew on each day 

```
"""Using value count to find aggregate trends and the percentage of flight delays"""
#Number of flights each carrier flew on each day
daily_flight = flightdata.pivot_table(index='flight_date', columns='unique_carrier', values='flight_num', aggfunc='count')
print(daily_flight)
daily_flight.plot(kind='line', figsize=[16,8], title='Daily flights by each carrier')
```

From the chart plotted, Carrier WN can be seen to fly more than the other carriers each day while Carrier UA is the least flown. Likewise, we can also see that on the 3rd, 6th and 10th January 2015, there was a drop in flights flown with the 10th seen to have a drastic drop in flights sale which is denoted in the chart as a dip.


ii.)	Sum of reasons for the flight delay

```
#Sum of reasons for the flight delay
delay_type = ['carrier_delay','weather_delay','late_aircraft_delay','nas_delay','security_delay']
daily_delay = flightdata.pivot_table(index='flight_date', values=delay_type, aggfunc='sum')
print(daily_delay)
daily_delay.plot(kind='line', figsize=[16,8], title='Sum in minutes of different delay reasons')
```

From the chart, 4th of January 2015 has the highest total number of delays compared to the other days with carrier_delay, late_aircraft_delay and nas_delay attaining its highest number. This significantly means, on the 4th of January, passengers experienced delay in carrier (maybe due to one technical fault or another), late arrival, boarding, or departure of aircrafts and likewise, delay on the part of the aviation services.  Security delay has the minimal value throughout the period of days not exceeding 950 minutes on the 3rd of January. The major source of delay in this given period was late_aircraft_delay which is usually high every day except from the 10th where we had a dip. 


Note, on the 10th of January 2015, is the day where we had drastic drop in flights which in turn caused a drop in flight delays.






iii.)Percentage of flight delays

```
#Percentage of flight delays
delayed_flights = flightdata['delayed'].value_counts()[1]
non_delayed_flights = flightdata['delayed']. value_counts()[0]
total_flights = delayed_flights + non_delayed_flights
delayed_percent = (delayed_flights / total_flights) * 100
print(delays)
print('Total flights:', total_flights)
print(round(float(delayed_percent), 1), '% of the total flights experienced flight delays.')
```

C. Numbers of flights were delayed longer than 20 minutes

```
"""Using lambda functions to find how many flights were delayed longer than 20 minutes"""
flightdata['20mins_flights_delay'] = flightdata['arr_delay'].apply(lambda x: x > 20)
num_of_delays = flightdata['20mins_flights_delay'].value_counts()
print(num_of_delays)
print('Flights delayed longer than 20 mins: ', num_of_delays[1])
```
